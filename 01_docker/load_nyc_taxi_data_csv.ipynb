Read the data from the csv file

```python
df = pd.read_csv("yellow_tripdata_2021-01.csv.gz", nrows=100)
```

convert dates into datetime datatypes

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

load sqlalchemy (requires install of sqlalchemy and psycopg2)

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

create SQL statement for new table schema

```python
print(pd.io.sql.get_schema(df, name="yellow_taxi_data"))
```

Not loading the whole data at once, but chunking and iterating to load.

```python
df_iter = pd.read_csv("yellow_tripdata_2021-01.csv.gz", iterator=True, chunksize=1e5)
```

next item will be retrieve using next()

```python
df = next(df_iter)
```

```python
len(df)
```

```python
df
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

Create empty table with header first

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

append next chunks

```python
%time df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

```python
from time import time
```

```python
while True:
    t_start = time()

    df = next(df_iter)
    
    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)

    df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print("next chunk inserted in %.3f seconds" % (t_end - t_start))
```
